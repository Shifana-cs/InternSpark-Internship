# Bias Analysis

The Fake News Detection model was trained on a dataset consisting mainly of political news articles.

Potential sources of bias:
- Political bias
- Source bias
- Temporal bias

The model may perform differently when applied to news from domains such as science, healthcare, sports, or entertainment.

Regular evaluation and dataset diversification are necessary to reduce bias and improve fairness.

# Responsible AI Recommendations

1. Use diverse datasets collected from multiple sources.
2. Continuously monitor model performance.
3. Retrain the model with recent data.
4. Use explainability techniques such as LIME.
5. Include human oversight for important decisions.

These practices help improve transparency, accountability, and trustworthiness in AI systems.
